```python
import yfinance as yf
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

btc = yf.download('BTC-USD', start='2024-01-01', end='2024-12-31')

X = btc["Close"].values

mean = X.mean()

std = X.std()

plt.plot(X, color="blue",  label="X")
plt.plot(np.full_like(X, mean), color="green", linestyle="--", label="Mean")
plt.plot(np.full_like(X, mean + std ), "r--", label="Upper Bond")
plt.plot(np.full_like(X, mean - std ), "r--", label="Lower Bond")

plt.legend()
plt.show()

sns.heatmap(btc.corr(), annot=True, cmap='coolwarm', fmt='.2f')

plt.show()

btc_upper=btc[X> mean + std]
btc_Lower=btc[X< mean - std ]

print('Min :' ,X.min())
print('max :' ,X.max())
print('mean :' ,X.mean())
print('std :' ,X.std())
print('Upper Bond :' ,mean + std)
print('Lower Bond :' ,mean - std  ,'\n\n\n\n')
print('لیست روزهایی که از حد بالای انحراف معیار بیشتر هستند',btc_upper.Close ,'\n\n\n\n')
print('لیست روزهایی که از حد پایین انحراف معیار کمتر هستند',btc_Lower.Close)
```
